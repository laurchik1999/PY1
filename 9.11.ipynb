```python
import math
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import numpy as np
```

```python
x_0 = 10
dt = 0.1
x_1 = x_0 - x_0 * dt
print ('x_1=', x_1)
x_2 = x_1 - x_1 * dt
print ('x_2=', x_2)
```

```python
tt = [0,]
xx = [x_0,]
final_t =  5
for i in range(int(final_t / dt)):
    x_last = xx[i]
    x_new = x_last - x_last * dt
    xx.append(x_new)
    t_new = tt[i]+dt
    tt.append(t_new)
plt.plot(tt,xx)
```

Уравнение маятника
$$ \frac{d\Theta^2}{dt^2} = - \sin\Theta $$ 

```python
theta0 = 45 * 3.14 / 180
omega0 = 0
t0 = 0
dt = 0.1
theta = [theta0,]
omega = [omega0,]
t = [t0,]
for i in range(500):
    omega_new = omega [i] - math.sin(theta[i])*dt
    omega.append(omega_new)
    theta_new = theta[i] + omega [i]*dt
    theta.append(theta_new)
    t_new = t[i] + dt
    t.append(t_new)
plt.plot(t,theta)    
```
